* Create histograms for quantitative variables. A histogram splits the (tabular) data into evenly sized intervals and displays the count of rows in each interval with bars. A histogram is similar to a bar chart, except that the "category" here is a range of values.

### Bar Chart 

__Bar Chart__ depicts the distribution of a categorical variable. In a bar chart, each level of the categorical variable is depicted with a bar, whose height indicates the frequency of data points that take on that level.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv("data/pokemon.csv")
print(pokemon.shape)
pokemon.head()
```

```python
sb.color_palette()
```

```python
base_color=sb.color_palette()[5]
```

```python
gen_order = pokemon['generation_id'].value_counts().index
```

```python
# Let's create a barchart on one of our variable 

sb.countplot(data=pokemon, x='generation_id', color=base_color, order=gen_order);
```

```python
# Now let's try with other variabl called type 1
base_color = sb.color_palette()[0]

sb.countplot(data = pokemon, x='type_1', color=base_color)
```

```python
# In the above image we are unable to read the names properly so let's try to make it readable 
base_color = sb.color_palette()[0]

sb.countplot(data = pokemon, x='type_1', color=base_color);
plt.xticks(rotation = 90); # this will rotate the label names  by 90 degrees on x axis
```
